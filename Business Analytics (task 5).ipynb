#### Cleaning the data to check for the Null values

```python
data.isnull().sum()
```

```python
Superstore = data.drop(['Ship Mode', 'Segment', 'Country', 'Postal Code' ], axis=1)
```

```python
Superstore.head()
```

#### Relationship Analysis

```python
corelation = Superstore.corr()
```

```python
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot = True)  
```

```python
sns.pairplot(Superstore)
```

#### Using Scatter Diagram 

```python
sns.relplot(x = 'Sales', y = 'Profit', hue='Category', data = Superstore)
```

```python
sns.relplot(x = 'Sales', y = 'Profit', hue='Sub-Category', data = Superstore)
```

```python
sns.relplot(x = 'Sales', y = 'Profit', hue='State', data = Superstore)
```

#### Using Histogram 

```python
sns.distplot(Superstore['Sales'])
```

```python
sns.distplot(Superstore['Discount'])
```

```python
sns.distplot(Superstore['Profit'])
```

```python
sns.distplot(Superstore['Quantity'])
```

#### Using Box Diagram

```python
sns.catplot(x= 'Sales', kind = 'box', data = Superstore)
```

```python
sns.catplot(x= 'Profit', kind = 'box', data = Superstore)
```
